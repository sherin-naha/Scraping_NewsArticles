# BBC Climate Change

```python
#base_url='https://www.bbc.com/'
url='https://www.bbc.co.uk/search?q=climate+change&page=1'
r1 = requests.get(url)
coverpage = r1.content
```

```python
soup1 = BeautifulSoup(coverpage, 'html5lib')
```

```python
news_title=[]
news_links=[]
type=[]
source_url=[]
for u in range(1,30):
    print(u)
    url='https://www.bbc.co.uk/search?q=climate+change&page='+str(u)
    print(url)
    r1 = requests.get(url)
    coverpage = r1.content
    soup1 = BeautifulSoup(coverpage, 'html5lib')
    
    
    coverpage_news = soup1.find_all('ul',role='list')[3].find_all('li')
    len(coverpage_news)
    
    for i in coverpage_news:
        #title
        news_title.append(i.find_all('a')[0].text)
        #link
        news_links.append(i.find_all('a')[0]['href'])
        #type
        type.append(i.find_all('span',type='category')[0].text)
        #source_url
        source_url.append(url)
```

```python
len(news_title)
```

```python
base_df=pd.DataFrame()
base_df['Headline']=news_title
base_df['Link']=news_links
base_df['Type']=type
base_df['source_url']=source_url
```

```python
base_df.head()
```

```python
base_df.to_csv('BBC_climate_Change_Archives.csv')
```

```python
df=base_df[base_df.Type=='News']
```

```python
df.reset_index(inplace=True)
```

```python
df['Tags']=''
df['Date']=''
df['Author']=''
df['News_body']=''
```

```python
df.head()
```

```python
author_list=[]
date_list=[]
tags_list=[]
body_list=[]
#scraping the news articles from each link
for i in range(len(df)):
    
    news_url=df['Link'][i]
    print(i)
    print(news_url)
    
    r1 = requests.get(news_url)
    newspage = r1.content
    soup2 = BeautifulSoup(newspage, 'html5lib')

    #get tags
    t=soup2.find('li',class_='tags-list__tags')
    if(t==None):
        tag=''
    else:
        tag=t.text

    #date
    d=soup2.find('div',class_='date date--v2')
    if(d==None):
        date=''
    else:
        date=d.text

    #author
    if(soup2.find('span',class_='byline__name')):
        author=soup2.find('span',class_='byline__name').text
    else:
        author=''

    #text
    text=''
    n=soup2.find('div',class_='story-body')
    if(n):
        for i in (n.find('div',class_='story-body__inner').find_all('p')):
            text=text+ (i.text.strip())
    
    author_list.append(author)
    date_list.append(date)
    tags_list.append(tag)
    body_list.append(text)
```

```python
df['Tags']=tags_list
df['Author']=author_list
df['News_body']=body_list
df['Date']=date_list
```

```python
df.head()
```

```python
df.to_csv('BBC_Climate change_archive_final.csv')
```
